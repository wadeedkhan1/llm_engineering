To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import ollama
from IPython.display import Markdown, display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
question = input("Enter your question: ")
```

```python
System_prompt = """
You are an assistant that is a technical expert, and gives detailed explanation for any questions asked.
"""
```

```python
User_prompt = f"Give a detailed explanation with relevant examples and analogies for the question:{question}"
```

```python
# Get Llama 3.2 to answer

response = ollama.chat(
        model=MODEL_LLAMA,
        messages=[
            {"role": "system", "content": System_prompt},
            {"role": "user", "content": User_prompt}
        ]
    )
```

```python
display(Markdown(response["message"]["content"]))
```

### Trying Gemini API

```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()
key = os.getenv('GOOGLE_API_KEY')

GEMINI_API_URL = "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent"
GEMINI_API_KEY = key
```

```python
response = requests.post(
        f"{GEMINI_API_URL}?key={GEMINI_API_KEY}",
        headers={'Content-Type': 'application/json'},
        json={
            "contents": [
            {"role": "user", "parts": [{"text": f"{System_prompt}\n\n{User_prompt}"}]}
            ]
        }
    )
```

```python
data = response.json()
reply_text = data["candidates"][0]["content"]["parts"][0]["text"]
display(Markdown(reply_text))
```
